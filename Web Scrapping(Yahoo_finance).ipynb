# Importing libraries

```python
import os
import numpy as np
import pandas as pd
import requests #to fetch data from webpage
from bs4 import BeautifulSoup #to beautify the data and noise removal
```

# Redirecting to url and request to fetch data

```python
#URL of webpage
url="https://in.finance.yahoo.com/quote/000001.SS/history?p=000001.SS"

#making request to fetch data from URL
response=requests.get(url)
```

```python
#content fetched from webpage
response.content
```

# Scrapping and removing noise

```python
#putting the fetched data in beautiful soup
soup=BeautifulSoup(response.content,"html.parser")
```

```python
soup
```

```python
#removing noise from fetched data
data=soup.find('table',{'class':'W(100%) M(0)'})
data=data.findAll('span')
data
```

```python
dat=data[7:-3]#data slicing
dat=[x.text for x in dat]#getting text from data
dat=np.array(dat).reshape(100,7)#creating array and arrange the data
dat
```

```python
cols=data[:7]
cols=[col.text for col in cols]
cols
```

# Creating the dataframe and storing the scrapped data into it

```python
df=pd.DataFrame(data=dat,columns=cols)
```

```python
df
```

```python
df.to_csv("yahoo_finance.csv",index=False)
```
